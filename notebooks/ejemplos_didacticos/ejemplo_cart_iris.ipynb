# Ejemplo didáctico: árbol CART con Iris

El conjunto Iris se utiliza para mostrar cómo un árbol de clasificación divide observaciones mediante reglas simples. Es un ejemplo didáctico independiente de la comparativa principal.


## Librerías y datos


```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
iris = load_iris(as_frame=True)
X = iris.data
y = iris.target

resumen_iris = pd.DataFrame([{
    "filas": X.shape[0],
    "variables": X.shape[1],
    "clases": len(iris.target_names),
}])

display(resumen_iris)
X.head()
```

## Entrenamiento de dos árboles

Se compara un árbol con profundidad limitada frente a otro sin limitar la profundidad. La versión limitada es más adecuada para visualizar el mecanismo de decisión.


```python
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)

arbol_limitado = DecisionTreeClassifier(max_depth=2, random_state=42)
arbol_completo = DecisionTreeClassifier(random_state=42)
arbol_limitado.fit(X_entrenamiento, y_entrenamiento)
arbol_completo.fit(X_entrenamiento, y_entrenamiento)

predicciones_limitado = arbol_limitado.predict(X_prueba)
predicciones_completo = arbol_completo.predict(X_prueba)

metricas = pd.DataFrame([
    {"Modelo": "CART max_depth=2", "Exactitud": accuracy_score(y_prueba, predicciones_limitado)},
    {"Modelo": "CART sin límite de profundidad", "Exactitud": accuracy_score(y_prueba, predicciones_completo)},
])

metricas
```

## Figura del árbol

La figura permite ver las primeras reglas de partición del árbol. Los resultados son solo ilustrativos y no forman parte de la evaluación principal.


```python
carpeta_figuras = Path("outputs/figures/ejemplos_didacticos")
carpeta_figuras.mkdir(parents=True, exist_ok=True)

fig, ax = plt.subplots(figsize=(11, 6))
plot_tree(
    arbol_limitado,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
    rounded=True,
    ax=ax,
)
ax.set_title("Árbol de clasificación didáctico con Iris")
fig.tight_layout()
fig.savefig(carpeta_figuras / "arbol_clasificacion_iris.png", dpi=180, bbox_inches="tight")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 4))
ConfusionMatrixDisplay.from_predictions(
    y_prueba,
    predicciones_limitado,
    display_labels=iris.target_names,
    cmap="Blues",
    ax=ax,
    colorbar=False,
)
ax.set_title("Matriz de confusión didáctica: Iris")
fig.tight_layout()
plt.show()
```
