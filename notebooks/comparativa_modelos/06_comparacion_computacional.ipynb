# Comparación computacional

Este cuaderno reúne los resultados generados por CART, Bagging y Boosting. No entrena modelos: solo lee las tablas ya generadas.


## Librerías


```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
```

## Carga de resultados


```python
cart_clasificacion = pd.read_csv("reports/tables/serie_computacional/cart_clasificacion_metricas.csv")
bagging_clasificacion = pd.read_csv("reports/tables/serie_computacional/bagging_clasificacion_metricas.csv")
boosting_clasificacion = pd.read_csv("reports/tables/serie_computacional/boosting_clasificacion_metricas.csv")

cart_regresion = pd.read_csv("reports/tables/serie_computacional/cart_regresion_metricas.csv")
bagging_regresion = pd.read_csv("reports/tables/serie_computacional/bagging_regresion_metricas.csv")
boosting_regresion = pd.read_csv("reports/tables/serie_computacional/boosting_regresion_metricas.csv")

resultados_clasificacion = pd.concat(
    [cart_clasificacion, bagging_clasificacion, boosting_clasificacion],
    ignore_index=True,
)
resultados_regresion = pd.concat(
    [cart_regresion, bagging_regresion, boosting_regresion],
    ignore_index=True,
)

display(resultados_clasificacion)
resultados_regresion
```

## Ordenación por métrica principal


```python
clasificacion_con_duration = (
    resultados_clasificacion[resultados_clasificacion["Escenario"] == "con duration"]
    .sort_values("F1 de la clase positiva", ascending=False)
)

clasificacion_sin_duration = (
    resultados_clasificacion[resultados_clasificacion["Escenario"] == "sin duration"]
    .sort_values("F1 de la clase positiva", ascending=False)
)

regresion_ordenada = resultados_regresion.sort_values("RMSE", ascending=True)

display(clasificacion_con_duration)
display(clasificacion_sin_duration)
regresion_ordenada
```

## Figuras y resumen


```python
Path("reports/tables/serie_computacional").mkdir(parents=True, exist_ok=True)
Path("reports/figures/serie_computacional").mkdir(parents=True, exist_ok=True)

clasificacion_con_duration.to_csv("reports/tables/serie_computacional/comparacion_clasificacion_con_duration.csv", index=False)
clasificacion_sin_duration.to_csv("reports/tables/serie_computacional/comparacion_clasificacion_sin_duration.csv", index=False)
regresion_ordenada.to_csv("reports/tables/serie_computacional/comparacion_regresion_consumo_electrico.csv", index=False)

fig, ax = plt.subplots(figsize=(10, 5))
grafico = resultados_clasificacion.copy()
grafico["Etiqueta"] = grafico["Modelo"] + "\n" + grafico["Escenario"]
ax.bar(grafico["Etiqueta"], grafico["F1 de la clase positiva"], color="#4c78a8")
ax.set_title("Comparación de clasificación: F1 de la clase positiva")
ax.set_ylabel("F1 de la clase positiva")
ax.tick_params(axis="x", rotation=60)
fig.tight_layout()
fig.savefig("reports/figures/serie_computacional/comparacion_clasificacion_f1.png", dpi=150)
plt.show()

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(regresion_ordenada["Modelo"], regresion_ordenada["RMSE"], color="#f58518")
ax.set_title("Comparación de regresión: RMSE")
ax.set_ylabel("RMSE")
ax.tick_params(axis="x", rotation=45)
fig.tight_layout()
fig.savefig("reports/figures/serie_computacional/comparacion_regresion_rmse.png", dpi=150)
plt.show()
```

## Mejores modelos según la métrica principal


```python
mejor_con_duration = clasificacion_con_duration.iloc[0]
mejor_sin_duration = clasificacion_sin_duration.iloc[0]
mejor_regresion = regresion_ordenada.iloc[0]

resumen = pd.DataFrame([
    {
        "Tarea": "clasificación con duration",
        "Modelo": mejor_con_duration["Modelo"],
        "Métrica principal": "F1 de la clase positiva",
        "Valor": mejor_con_duration["F1 de la clase positiva"],
    },
    {
        "Tarea": "clasificación sin duration",
        "Modelo": mejor_sin_duration["Modelo"],
        "Métrica principal": "F1 de la clase positiva",
        "Valor": mejor_sin_duration["F1 de la clase positiva"],
    },
    {
        "Tarea": "regresión consumo eléctrico",
        "Modelo": mejor_regresion["Modelo"],
        "Métrica principal": "RMSE",
        "Valor": mejor_regresion["RMSE"],
    },
])

resumen.to_csv("reports/tables/serie_computacional/comparacion_resumen_mejores_modelos.csv", index=False)
resumen
```
